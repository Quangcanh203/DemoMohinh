```python
# -*- coding: utf-8 -*-
"""
Mã nguồn huấn luyện ResNet50 sử dụng dữ liệu đã được tiền xử lý và lưu dưới dạng .npy
từ đoạn mã "chay_doan_code_du_lieu_nay.py"
"""

from google.colab import drive
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization, Input
from tensorflow.keras.regularizers import l2

# 1. MOUNT GOOGLE DRIVE
drive.mount('/content/gdrive', force_remount=True)

# 2. ĐỊNH NGHĨA CÁC ĐƯỜNG DẪN VÀ THAM SỐ
BASE_DRIVE_PATH = '/content/gdrive/My Drive/chest_xray'
NEW_DATASET_ROOT_NAME_FROM_PREVIOUS_CODE = 'chest_xray_phan_chia_moi'
NUMPY_DATA_DIR_NAME_FROM_PREVIOUS_CODE = 'numpy_data'
MODEL_SAVE_SUBDIR = 'resnet50_trained_models_v4' # Đổi tên thư mục để tránh ghi đè

# Đường dẫn đến thư mục chứa các file .npy
numpy_data_dir = os.path.join(BASE_DRIVE_PATH, NEW_DATASET_ROOT_NAME_FROM_PREVIOUS_CODE, NUMPY_DATA_DIR_NAME_FROM_PREVIOUS_CODE)

# Đường dẫn để lưu các mô hình đã huấn luyện
model_save_path_root = os.path.join(BASE_DRIVE_PATH, NEW_DATASET_ROOT_NAME_FROM_PREVIOUS_CODE, MODEL_SAVE_SUBDIR)

if not os.path.exists(model_save_path_root):
    os.makedirs(model_save_path_root)
    print(f"Đã tạo thư mục lưu mô hình: {model_save_path_root}")
else:
    print(f"Thư mục lưu mô hình đã tồn tại: {model_save_path_root}")

IMG_SIZE = 224
NUM_CLASSES = 2
CLASS_LABELS = ['NORMAL', 'PNEUMONIA']

# 3. TẢI DỮ LIỆU TỪ CÁC FILE .NPY
print(f"\nĐang tải dữ liệu NumPy từ: {numpy_data_dir}")
try:
    X_train_orig = np.load(os.path.join(numpy_data_dir, "X_train.npy"))
    y_train_orig = np.load(os.path.join(numpy_data_dir, "y_train.npy"))
    X_val_orig = np.load(os.path.join(numpy_data_dir, "X_val.npy"))
    y_val_orig = np.load(os.path.join(numpy_data_dir, "y_val.npy"))
    X_test_final = np.load(os.path.join(numpy_data_dir, "X_test.npy"))
    y_test_final = np.load(os.path.join(numpy_data_dir, "y_test.npy"))
except FileNotFoundError as e:
    print(f"LỖI: Không tìm thấy tệp dữ liệu NumPy. Vui lòng đảm bảo đoạn mã xử lý dữ liệu đầu tiên đã chạy thành công.")
    print(e)
    exit()

print(f"Kích thước X_train_orig: {X_train_orig.shape}, y_train_orig: {y_train_orig.shape}")
print(f"Kích thước X_val_orig: {X_val_orig.shape}, y_val_orig: {y_val_orig.shape}")
print(f"Kích thước X_test_final: {X_test_final.shape}, y_test_final: {y_test_final.shape}")

# Chuẩn hóa dữ liệu
X_train_orig = X_train_orig.astype('float32') / 255.0
X_val_orig = X_val_orig.astype('float32') / 255.0
X_test_final = X_test_final.astype('float32') / 255.0

# Gộp dữ liệu train và val gốc để chuẩn bị cho K-Fold và tập warm-up
X_for_kfold_and_warmup = np.concatenate((X_train_orig, X_val_orig), axis=0)
y_for_kfold_and_warmup = np.concatenate((y_train_orig, y_val_orig), axis=0)

print(f"\nKích thước dữ liệu gộp (X_train_orig + X_val_orig) cho K-Fold và chia warm-up:")
print(f"X_for_kfold_and_warmup: {X_for_kfold_and_warmup.shape}, y_for_kfold_and_warmup: {y_for_kfold_and_warmup.shape}")

# Chia tập dữ liệu gộp thành tập train/val cho giai đoạn WARM-UP
X_train_warmup, X_val_warmup, y_train_warmup, y_val_warmup = train_test_split(
    X_for_kfold_and_warmup, y_for_kfold_and_warmup,
    test_size=0.2, # 20% cho validation trong warm-up
    random_state=42,
    stratify=y_for_kfold_and_warmup
)

print(f"\nKích thước dữ liệu cho WARM-UP:")
print(f"X_train_warmup: {X_train_warmup.shape}, y_train_warmup: {y_train_warmup.shape}")
print(f"X_val_warmup: {X_val_warmup.shape}, y_val_warmup: {y_val_warmup.shape}")

# Dữ liệu cho K-Fold là toàn bộ (X_train_orig + X_val_orig)
X_kfold_data = X_for_kfold_and_warmup
y_kfold_data = y_for_kfold_and_warmup

# 4. HÀM TẠO MÔ HÌNH
def create_resnet_model(input_shape=(IMG_SIZE, IMG_SIZE, 3), num_classes=NUM_CLASSES):
    # Sử dụng một Input layer tường minh
    input_tensor = Input(shape=input_shape, name="input_layer_explicit")

    # base_model sẽ được coi là một khối, không phải là một layer lồng nhau
    # khi dùng cách này để tạo model thì các layer con của base_model sẽ được trải phẳng.
    base_model_instance = ResNet50(weights='imagenet', include_top=False, input_tensor=input_tensor, pooling=None)
    # Đặt tên cho output của base_model để dễ tham chiếu nếu cần sau này (ít quan trọng hơn)
    # base_model_output = tf.keras.layers.Layer(name="resnet50_output_node")(base_model_instance.output)

    # Đặt base_model.trainable = False ban đầu.
    # Thuộc tính trainable của base_model_instance sẽ ảnh hưởng đến các lớp con của nó.
    base_model_instance.trainable = False

    x = base_model_instance.output # Lấy output từ base_model
    x = GlobalAveragePooling2D(name="global_avg_pooling")(x)
    x = BatchNormalization(name="bn_after_gap_1")(x)
    x = Dense(512, activation='relu', kernel_regularizer=l2(0.001), name="dense_512")(x)
    x = Dropout(0.5, name="dropout_layer")(x)
    x = BatchNormalization(name="bn_after_dense_1")(x)
    outputs = Dense(num_classes, activation='softmax', name="output_dense")(x)

    # Model được tạo từ input_tensor ban đầu và output cuối cùng
    model = Model(inputs=input_tensor, outputs=outputs)
    return model

# 5. HÀM VẼ BIỂU ĐỒ HUẤN LUYỆN
def plot_training_history(history, title_prefix=""):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(len(acc))

    plt.figure(figsize=(18, 6))

    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, 'royalblue', label='Train Accuracy')
    plt.plot(epochs_range, val_acc, 'darkorange', linestyle='--', label='Validation Accuracy')
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Accuracy', fontsize=12)
    plt.title(f'{title_prefix} Accuracy', fontsize=14)
    plt.legend(loc='lower right')
    plt.grid(True, linestyle='--', alpha=0.7)

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, 'royalblue', label='Train Loss')
    plt.plot(epochs_range, val_loss, 'darkorange', linestyle='--', label='Validation Loss')
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Loss', fontsize=12)
    plt.title(f'{title_prefix} Loss', fontsize=14)
    plt.legend(loc='upper right')
    plt.grid(True, linestyle='--', alpha=0.7)

    plt.suptitle(f'{title_prefix} Training History', fontsize=16, fontweight='bold')
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

# 6. CALLBACKS
lr_scheduler_warmup = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6, verbose=1)
early_stopping_warmup = EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True, verbose=1)

lr_scheduler_finetune = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=4, min_lr=1e-7, verbose=1)
early_stopping_finetune = EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True, verbose=1)

# 7. GIAI ĐOẠN WARM-UP
print("\n--- BẮT ĐẦU GIAI ĐOẠN WARM-UP ---")
model_for_warmup = create_resnet_model()

# Trong warm-up, base_model_instance.trainable đã được đặt là False trong create_resnet_model
# Điều này làm cho tất cả các lớp của ResNet50 không thể huấn luyện.
# Các lớp custom (Dense, Dropout, BN) mặc định là trainable=True và sẽ được huấn luyện.

model_for_warmup.compile(
    loss='sparse_categorical_crossentropy',
    optimizer=Adam(learning_rate=1e-3),
    metrics=['accuracy']
)

print("Trạng thái trainable của các lớp TRƯỚC WARM-UP (kiểm tra các lớp ResNet):")
resnet_layers_found = False
for layer in model_for_warmup.layers:
    # Tên các lớp ResNet thường chứa "conv", "bn", "relu", "add", "pool"
    # và không phải là các lớp custom bạn đã thêm (global_avg_pooling, bn_after_gap_1, etc.)
    if layer.name.startswith("conv") or layer.name.startswith("pool") or \
       layer.name.startswith("res") or layer.name.startswith("bn") and \
       not layer.name in ["bn_after_gap_1", "bn_after_dense_1"]: # Loại trừ BN custom
        print(f"  - {layer.name}: {layer.trainable}")
        resnet_layers_found = True
    elif layer.name in ["global_avg_pooling", "bn_after_gap_1", "dense_512", "dropout_layer", "bn_after_dense_1", "output_dense", "input_layer_explicit"]:
        print(f" - {layer.name} (Custom/Input): {layer.trainable}")
if not resnet_layers_found:
    print("  (Không tìm thấy lớp nào được nhận diện là thuộc ResNet theo tên)")


history_warmup = model_for_warmup.fit(
    X_train_warmup, y_train_warmup,
    validation_data=(X_val_warmup, y_val_warmup),
    epochs=20,
    batch_size=32,
    callbacks=[lr_scheduler_warmup, early_stopping_warmup],
    verbose=1
)

warmup_model_save_path = os.path.join(model_save_path_root, "ResNet50_WarmUp_ChestXRay.keras")
model_for_warmup.save(warmup_model_save_path)
print(f"Đã lưu mô hình sau Warm-up tại: {warmup_model_save_path}")

print("\nHiển thị biểu đồ huấn luyện cho giai đoạn Warm-up:")
plot_training_history(history_warmup, title_prefix="Warm-up Phase")


# 8. GIAI ĐOẠN FINE-TUNING VỚI K-FOLD
print("\n--- BẮT ĐẦU GIAI ĐOẠN FINE-TUNING VỚI K-FOLD ---")
k_folds = 5
skf = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=42)

all_fold_histories = []
fold_no = 1

for train_idx, val_idx in skf.split(X_kfold_data, y_kfold_data):
    print(f'\n========== Fold {fold_no}/{k_folds} ==========')
    X_train_fold, X_val_fold = X_kfold_data[train_idx], X_kfold_data[val_idx]
    y_train_fold, y_val_fold = y_kfold_data[train_idx], y_kfold_data[val_idx]

    print(f"Kích thước dữ liệu cho Fold {fold_no}:")
    print(f"X_train_fold: {X_train_fold.shape}, y_train_fold: {y_train_fold.shape}")
    print(f"X_val_fold: {X_val_fold.shape}, y_val_fold: {y_val_fold.shape}")

    model_for_finetune = tf.keras.models.load_model(warmup_model_save_path)

    # --- ĐIỀU CHỈNH LOGIC UNFREEZING ---
    # Xác định điểm kết thúc của các lớp ResNet (ngay trước GlobalAveragePooling2D)
    idx_before_gap = -1
    for i, layer in enumerate(model_for_finetune.layers):
        if layer.name == "global_avg_pooling":
            idx_before_gap = i -1
            break
    if idx_before_gap == -1:
        raise ValueError("Không tìm thấy lớp 'global_avg_pooling'. Kiểm tra tên lớp.")

    print(f"Lớp cuối cùng của phần ResNet được xác định là: {model_for_finetune.layers[idx_before_gap].name} (index {idx_before_gap})")

    # Chọn lớp để bắt đầu fine-tuning (ví dụ: từ block cuối cùng)
    fine_tune_at_layer_name = 'conv5_block1_conv' # Hoặc conv4_block1_conv
    # Lưu ý: Tên chính xác có thể là 'conv5_block1_1_conv' nếu ResNet được mở rộng.
    # Hãy kiểm tra tên bằng model_for_finetune.summary() nếu không chắc.
    # Một cách khác là unfreeze một số lượng block nhất định từ cuối.
    # Ví dụ: để unfreeze block 5, bạn cần tìm 'conv5_block1_...'
    # Ví dụ: để unfreeze block 4 và 5, bạn cần tìm 'conv4_block1_...'

    # Cố gắng tìm tên lớp chính xác hơn (ResNet50 có các shortcut connections với tên khác)
    # Ví dụ, 'conv5_block1_0_conv' là lớp conv trong shortcut path của block đầu tiên của stage 5
    # 'conv5_block1_1_conv' là lớp conv đầu tiên trong main path của block đầu tiên của stage 5
    # Chúng ta thường muốn bắt đầu fine-tune từ lớp đầu tiên của một block.
    # Cách an toàn hơn là đếm ngược số block.
    # ResNet50 có các stage conv2, conv3, conv4, conv5.
    # Chúng ta muốn fine-tune từ conv5 hoặc conv4.
    # Block conv5 bắt đầu với các lớp có tên chứa 'conv5_block1'.

    fine_tune_start_index = -1
    # Tìm chỉ số của lớp có tên fine_tune_at_layer_name
    # Hoặc tìm chỉ số của lớp có tên chứa "conv5_block1" (hoặc "conv4_block1")
    # Chú ý: tên có thể khác nhau giữa các phiên bản TF/Keras hoặc nếu base model được tạo khác đi.

    # Chiến lược: Unfreeze các lớp từ một block cụ thể trở đi.
    # Ví dụ: unfreeze tất cả các lớp thuộc 'conv5_...'
    target_block_prefix = "conv5_block" # Unfreeze từ block 5
    # target_block_prefix = "conv4_block" # Unfreeze từ block 4

    for i, layer in enumerate(model_for_finetune.layers):
        if layer.name.startswith(target_block_prefix):
            if fine_tune_start_index == -1: # Chỉ lấy chỉ số của lớp đầu tiên khớp
                 fine_tune_start_index = i
                 print(f"Fine-tuning sẽ bắt đầu từ lớp: {layer.name} (index {i}) dựa trên prefix '{target_block_prefix}'")
            # Đặt trainable = True cho các lớp thuộc block mục tiêu trở đi, VÀ là một phần của ResNet
            if i <= idx_before_gap : # Đảm bảo lớp này là một phần của ResNet gốc
                if isinstance(layer, tf.keras.layers.BatchNormalization):
                    layer.trainable = False # BN của pretrained model ở inference mode
                else:
                    layer.trainable = True
        elif i <= idx_before_gap: # Các lớp ResNet trước block mục tiêu
             layer.trainable = False
        else: # Các lớp custom thêm vào sau ResNet (GAP, Dense, BN custom, Dropout, Output)
            if isinstance(layer, tf.keras.layers.BatchNormalization) and layer.name in ["bn_after_gap_1", "bn_after_dense_1"]:
                layer.trainable = True # BN custom nên được train
            elif not isinstance(layer, tf.keras.layers.BatchNormalization): # Dense, Dropout, GAP, Output
                layer.trainable = True
            # else: BN của pretrain đã được xử lý ở trên


    if fine_tune_start_index == -1:
        print(f"LƯU Ý: Không tìm thấy lớp nào bắt đầu với prefix '{target_block_prefix}'. Kiểm tra tên lớp.")
        print(f"Sẽ thử unfreeze 20 lớp cuối cùng của phần ResNet.")
        num_resnet_layers_to_unfreeze = 20
        fine_tune_start_index = max(0, idx_before_gap - num_resnet_layers_to_unfreeze + 1)
        print(f"Fine-tuning sẽ bắt đầu từ lớp: {model_for_finetune.layers[fine_tune_start_index].name} (index {fine_tune_start_index})")
        for i in range(fine_tune_start_index, idx_before_gap + 1):
            layer = model_for_finetune.layers[i]
            if isinstance(layer, tf.keras.layers.BatchNormalization):
                layer.trainable = False
            else:
                layer.trainable = True

    # Đảm bảo các lớp custom cuối cùng là trainable
    for layer_name in ["global_avg_pooling", "bn_after_gap_1", "dense_512", "dropout_layer", "bn_after_dense_1", "output_dense"]:
        try:
            custom_layer = model_for_finetune.get_layer(name=layer_name)
            if isinstance(custom_layer, tf.keras.layers.BatchNormalization):
                 custom_layer.trainable = True # BN custom phải train
            else:
                 custom_layer.trainable = True
        except ValueError:
            print(f"Cảnh báo: Không tìm thấy lớp custom {layer_name} khi đặt trainable.")


    model_for_finetune.compile(
        loss='sparse_categorical_crossentropy',
        optimizer=Adam(learning_rate=1e-5),
        metrics=['accuracy']
    )

    print(f"Trạng thái trainable của các lớp cho Fine-tuning Fold {fold_no} (SAU KHI CÀI ĐẶT):")
    for i, layer in enumerate(model_for_finetune.layers):
        print(f" - {layer.name}: {layer.trainable} (Thuộc ResNet gốc: {i <= idx_before_gap})")


    print(f"Bắt đầu huấn luyện Fold {fold_no} (Fine-tuning)...")
    history_fold = model_for_finetune.fit(
        X_train_fold, y_train_fold,
        validation_data=(X_val_fold, y_val_fold),
        epochs=30,
        batch_size=32,
        callbacks=[lr_scheduler_finetune, early_stopping_finetune],
        verbose=1
    )

    all_fold_histories.append(history_fold.history)

    fold_model_save_path = os.path.join(model_save_path_root, f"ResNet50_FineTune_ChestXRay_Fold{fold_no}.keras")
    model_for_finetune.save(fold_model_save_path)
    print(f"Đã lưu mô hình cho Fold {fold_no} tại: {fold_model_save_path}")

    print(f"\nĐánh giá Fold {fold_no}:")
    y_pred_val_fold_probs = model_for_finetune.predict(X_val_fold)
    y_pred_val_fold = np.argmax(y_pred_val_fold_probs, axis=1)
    val_accuracy_fold = accuracy_score(y_val_fold, y_pred_val_fold)
    print(f'Fold {fold_no} - Validation Accuracy: {val_accuracy_fold:.4f}')

    print(f"\nHiển thị biểu đồ huấn luyện cho Fold {fold_no}:")
    plot_training_history(history_fold, title_prefix=f"Fine-tuning Fold {fold_no}")

    cm_val_fold = confusion_matrix(y_val_fold, y_pred_val_fold)
    plt.figure(figsize=(7, 5))
    sns.heatmap(cm_val_fold, annot=True, fmt='d', cmap='Blues', xticklabels=CLASS_LABELS, yticklabels=CLASS_LABELS)
    plt.xlabel('Dự đoán (Predicted Label)')
    plt.ylabel('Thực tế (True Label)')
    plt.title(f'Confusion Matrix - Validation Set (Fold {fold_no})')
    plt.show()
    print(f"Classification Report - Validation Set (Fold {fold_no}):")
    print(classification_report(y_val_fold, y_pred_val_fold, target_names=CLASS_LABELS, zero_division=0))

    fold_no += 1

print("\nĐã hoàn thành huấn luyện K-Fold và lưu các mô hình!")

# 9. CHỌN MÔ HÌNH TỐT NHẤT TỪ K-FOLD VÀ ĐÁNH GIÁ TRÊN TẬP TEST
if all_fold_histories:
    best_fold_index = -1
    best_val_accuracy = -1.0

    for i, history_data in enumerate(all_fold_histories):
        current_fold_best_val_acc = np.max(history_data['val_accuracy'])
        print(f"Fold {i+1} - Best Validation Accuracy in its history: {current_fold_best_val_acc:.4f}")

        if current_fold_best_val_acc > best_val_accuracy:
            best_val_accuracy = current_fold_best_val_acc
            best_fold_index = i

    if best_fold_index != -1:
        best_fold_number = best_fold_index + 1
        print(f"\nFold tốt nhất dựa trên Validation Accuracy là Fold {best_fold_number} với Val Accuracy: {best_val_accuracy:.4f}")
        best_model_path = os.path.join(model_save_path_root, f"ResNet50_FineTune_ChestXRay_Fold{best_fold_number}.keras")

        if os.path.exists(best_model_path):
            print(f"\n--- ĐÁNH GIÁ TRÊN TẬP TEST SỬ DỤNG MÔ HÌNH TỪ FOLD {best_fold_number} ---")
            model_best_fold = tf.keras.models.load_model(best_model_path, compile=True)

            y_pred_test_probs = model_best_fold.predict(X_test_final)
            y_pred_test = np.argmax(y_pred_test_probs, axis=1)

            test_accuracy = accuracy_score(y_test_final, y_pred_test)
            print(f'Test Accuracy (Model from Fold {best_fold_number}): {test_accuracy:.4f}')

            cm_test = confusion_matrix(y_test_final, y_pred_test)
            plt.figure(figsize=(7, 5))
            sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues', xticklabels=CLASS_LABELS, yticklabels=CLASS_LABELS)
            plt.xlabel('Dự đoán (Predicted Label)')
            plt.ylabel('Thực tế (True Label)')
            plt.title(f'Confusion Matrix - Test Set (Model from Fold {best_fold_number})')
            plt.show()
            print(f"Classification Report - Test Set (Model from Fold {best_fold_number}):")
            print(classification_report(y_test_final, y_pred_test, target_names=CLASS_LABELS, zero_division=0))
        else:
            print(f"LỖI: Không tìm thấy mô hình tốt nhất tại: {best_model_path} để đánh giá trên tập test.")
    else:
        print("Không thể xác định mô hình tốt nhất từ K-Fold.")
else:
    print("Không có lịch sử huấn luyện K-Fold nào được ghi lại để chọn mô hình tốt nhất.")

print("\n--- TOÀN BỘ QUÁ TRÌNH HUẤN LUYỆN VÀ ĐÁNH GIÁ HOÀN TẤT ---")
```
