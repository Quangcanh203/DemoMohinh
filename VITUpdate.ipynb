```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils import class_weight
import cv2

print(f"Phiên bản TensorFlow đang sử dụng: {tf.__version__}")

BASE_PATH_PROCESSED_DATA = '/content/gdrive/My Drive/chest_xray/chest_xray_phan_chia_moi'
NUMPY_SAVE_DIR_NAME = 'numpy_data'
NUMPY_DATA_DIR = os.path.join(BASE_PATH_PROCESSED_DATA, NUMPY_SAVE_DIR_NAME)
MODEL_SAVE_PATH = 'best_vit_chest_xray_model_v4_patch16.keras'

IMG_SIZE = 224
NUM_CLASSES = 2
PATCH_SIZE = 16
PROJECTION_DIM = 256
TRANSFORMER_LAYERS = 8
NUM_HEADS = 8
MLP_DIM_TRANSFORMER_ENCODER = PROJECTION_DIM * 4
NUM_PATCHES = (IMG_SIZE // PATCH_SIZE) ** 2

LEARNING_RATE = 1e-4
WEIGHT_DECAY = 5e-5
BATCH_SIZE = 16
EPOCHS = 150
PATIENCE_EARLY_STOPPING = 30
APPLY_DATA_AUGMENTATION = True
APPLY_CLASS_WEIGHTS = True
MANUAL_CLASS_WEIGHTS = False
CUSTOM_WEIGHT_NORMAL = 0.8
CUSTOM_WEIGHT_PNEUMONIA = 1.2
LABEL_SMOOTHING = 0.0

def preprocess_images(images,
                      target_size=(IMG_SIZE, IMG_SIZE),
                      apply_blur=True, blur_ksize=(5,5),
                      apply_brightness=True, brightness_factor=1.2,
                      apply_sharpen=True):
    processed_images = []
    for img in images:
        img_resized = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)
        if len(img_resized.shape) == 2:
            img_resized = cv2.cvtColor(img_resized, cv2.COLOR_GRAY2RGB)
        if apply_blur:
            img_blurred = cv2.GaussianBlur(img_resized, blur_ksize, 0)
        else:
            img_blurred = img_resized
        if apply_brightness:
            img_bright = cv2.convertScaleAbs(img_blurred, alpha=brightness_factor, beta=0)
        else:
            img_bright = img_blurred
        if apply_sharpen:
            # Kernel sharpen
            kernel = np.array([[0, -1, 0],
                               [-1, 5,-1],
                               [0, -1, 0]])
            img_sharp = cv2.filter2D(img_bright, -1, kernel)
        else:
            img_sharp = img_bright
        processed_images.append(img_sharp)
    return np.array(processed_images)

try:
    from google.colab import drive
    drive.mount('/content/gdrive', force_remount=True)
    print(f"Đã mount Google Drive. Thư mục dữ liệu NumPy dự kiến: {NUMPY_DATA_DIR}")
    if 'google.colab' in str(get_ipython()):
        MODEL_SAVE_PATH = os.path.join('/content', MODEL_SAVE_PATH)
    print(f"Mô hình tốt nhất sẽ được lưu tại: {MODEL_SAVE_PATH}")
except Exception as e:
    print(f"Lỗi khi mount Google Drive hoặc xác định get_ipython: {e}")
    try:
        if 'google.colab' not in str(get_ipython()):
             MODEL_SAVE_PATH = os.path.join(os.getcwd(), MODEL_SAVE_PATH)
    except NameError:
        MODEL_SAVE_PATH = os.path.join(os.getcwd(), MODEL_SAVE_PATH)
    print(f"Mô hình tốt nhất sẽ được lưu tại: {MODEL_SAVE_PATH}")

print("\n--- TẢI DỮ LIỆU NUMPY ---")
if not os.path.exists(NUMPY_DATA_DIR):
    print(f"LỖI: Thư mục {NUMPY_DATA_DIR} không tồn tại. Vui lòng kiểm tra lại đường dẫn.")
    exit()
X_train = np.load(os.path.join(NUMPY_DATA_DIR, "X_train.npy"))
y_train = np.load(os.path.join(NUMPY_DATA_DIR, "y_train.npy"))
X_val = np.load(os.path.join(NUMPY_DATA_DIR, "X_val.npy"))
y_val = np.load(os.path.join(NUMPY_DATA_DIR, "y_val.npy"))
X_test = np.load(os.path.join(NUMPY_DATA_DIR, "X_test.npy"))
y_test = np.load(os.path.join(NUMPY_DATA_DIR, "y_test.npy"))
print(f"Shape X_train: {X_train.shape}, y_train: {y_train.shape}")

print("\n--- TIỀN XỬ LÝ ẢNH: Resize, Blur, Brightness ---")
X_train_preprocessed = preprocess_images(X_train, apply_sharpen=True)
X_val_preprocessed = preprocess_images(X_val, apply_sharpen=True)
X_test_preprocessed = preprocess_images(X_test, apply_sharpen=True)

print("Ảnh đã được resize, làm mờ và tăng sáng.")

import matplotlib.pyplot as plt

def show_sample_images(before_norm, after_norm, num_images=2):
    plt.figure(figsize=(12, 5))
    for i in range(num_images):
        # Ảnh sau tiền xử lý nhưng chưa chuẩn hóa
        plt.subplot(2, num_images, i + 1)
        plt.imshow(before_norm[i].astype(np.uint8))
        plt.title(f"Before normalization #{i+1}")
        plt.axis('off')

        # Ảnh sau chuẩn hóa pixel
        plt.subplot(2, num_images, num_images + i + 1)
        img_show = after_norm[i]
        plt.imshow(img_show)
        plt.title(f"After normalization #{i+1}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

print("\n--- HIỂN THỊ ẢNH MẪU TRƯỚC VÀ SAU CHUẨN HÓA ---")
show_sample_images(X_train_preprocessed, X_train_preprocessed.astype("float32") / 255.0)
print("--- TIẾP TỤC BƯỚC SAU ---")

# 4. Chuẩn hóa pixel
print("\n--- CHUẨN HÓA GIÁ TRỊ PIXEL ---")
X_train = X_train_preprocessed.astype("float32") / 255.0
X_val = X_val_preprocessed.astype("float32") / 255.0
X_test = X_test_preprocessed.astype("float32") / 255.0
print("Đã chuẩn hóa giá trị pixel.")

class Patches(layers.Layer):
    def __init__(self, patch_size, **kwargs):
        super().__init__(**kwargs)
        self.patch_size = patch_size
    def call(self, images):
        batch_size = tf.shape(images)[0]
        patches = tf.image.extract_patches(
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1,1,1,1],
            padding="VALID",
        )
        patch_dims = patches.shape[-1]
        patches = tf.reshape(patches, [batch_size, -1, patch_dims])
        return patches
    def get_config(self):
        config = super().get_config()
        config.update({"patch_size": self.patch_size})
        return config

class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim, **kwargs):
        super().__init__(**kwargs)
        self.num_patches = num_patches
        self.projection = layers.Dense(projection_dim)
        self.position_embedding = layers.Embedding(input_dim=num_patches, output_dim=projection_dim)
    def call(self, patch):
        positions = tf.range(start=0, limit=self.num_patches, delta=1)
        projected = self.projection(patch)
        return projected + self.position_embedding(positions)
    def get_config(self):
        config = super().get_config()
        config.update({"num_patches": self.num_patches, "projection_dim": self.projection.units})
        return config

def mlp_block(x, hidden_units, dropout_rate, transformer_layer_index, block_name_prefix="mlp_block"):
    for idx, units in enumerate(hidden_units):
        x = layers.Dense(units, activation=tf.nn.gelu,
                         name=f"{block_name_prefix}_transformer_{transformer_layer_index}_dense_{idx}")(x)
        x = layers.Dropout(dropout_rate,
                          name=f"{block_name_prefix}_transformer_{transformer_layer_index}_dropout_{idx}")(x)
    return x

def create_vit_classifier(input_shape, patch_size, num_patches, projection_dim,
                          transformer_layers, num_heads, mlp_dim_transformer_encoder,
                          num_classes, apply_data_augmentation=False, mlp_head_units_config=None):
    inputs = layers.Input(shape=input_shape)
    x = inputs
    if apply_data_augmentation:
        aug = keras.Sequential([
            layers.RandomFlip("horizontal"),
            layers.RandomRotation(0.05),
            layers.RandomZoom(0.05, 0.05)
        ])
        x = aug(x)
    patches = Patches(patch_size)(x)
    encoded = PatchEncoder(num_patches, projection_dim)(patches)
    for i in range(transformer_layers):
        x1 = layers.LayerNormalization(epsilon=1e-6)(encoded)
        key_dim = projection_dim // num_heads
        attn_output = layers.MultiHeadAttention(num_heads=num_heads, key_dim=key_dim, dropout=0.1)(x1, x1)
        x2 = layers.Add()([attn_output, encoded])
        x3 = layers.LayerNormalization(epsilon=1e-6)(x2)
        x3 = mlp_block(x3, [mlp_dim_transformer_encoder, projection_dim], 0.1, i)
        encoded = layers.Add()([x3, x2])
    representation = layers.LayerNormalization(epsilon=1e-6)(encoded)
    representation = layers.GlobalAveragePooling1D()(representation)
    features = representation
    if mlp_head_units_config is None:
        mlp_head_units_config = [projection_dim * 2, projection_dim]
    for idx, units in enumerate(mlp_head_units_config):
        features = layers.Dense(units, activation=tf.nn.gelu)(features)
        features = layers.Dropout(0.4)(features)
    logits = layers.Dense(num_classes, activation="softmax")(features)
    return keras.Model(inputs=inputs, outputs=logits, name="vit_classifier_patch16")

vit_classifier = create_vit_classifier(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    patch_size=PATCH_SIZE,
    num_patches=NUM_PATCHES,
    projection_dim=PROJECTION_DIM,
    transformer_layers=TRANSFORMER_LAYERS,
    num_heads=NUM_HEADS,
    mlp_dim_transformer_encoder=MLP_DIM_TRANSFORMER_ENCODER,
    num_classes=NUM_CLASSES,
    apply_data_augmentation=APPLY_DATA_AUGMENTATION
)
vit_classifier.summary()

# Cập nhật learning rate schedule
steps_per_epoch = max(1, len(X_train)//BATCH_SIZE)
total_decay_steps = EPOCHS * steps_per_epoch
lr_schedule = tf.keras.optimizers.schedules.CosineDecay(
    initial_learning_rate=LEARNING_RATE,
    decay_steps=total_decay_steps,
    alpha=0.0
)
optimizer = tf.keras.optimizers.AdamW(learning_rate=lr_schedule, weight_decay=WEIGHT_DECAY)

loss_function = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False)

vit_classifier.compile(optimizer=optimizer, loss=loss_function, metrics=["accuracy"])

early_stopping = keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=PATIENCE_EARLY_STOPPING, restore_best_weights=True)
model_checkpoint = keras.callbacks.ModelCheckpoint(MODEL_SAVE_PATH, monitor='val_accuracy', save_best_only=True)

class_weight_dict = None
if APPLY_CLASS_WEIGHTS:
    if MANUAL_CLASS_WEIGHTS:
        class_weight_dict = {0: CUSTOM_WEIGHT_NORMAL, 1: CUSTOM_WEIGHT_PNEUMONIA}
    else:
        classes, counts = np.unique(y_train, return_counts=True)
        weights = class_weight.compute_class_weight('balanced', classes=classes, y=y_train)
        class_weight_dict = dict(zip(classes, weights))

history = vit_classifier.fit(
    X_train, y_train,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=(X_val, y_val),
    callbacks=[early_stopping, model_checkpoint],
    class_weight=class_weight_dict,
    verbose=1
)

# Load best model and evaluate
try:
    best_model = keras.models.load_model(MODEL_SAVE_PATH, custom_objects={'Patches': Patches, 'PatchEncoder': PatchEncoder})
except Exception:
    best_model = vit_classifier

loss, acc = best_model.evaluate(X_test, y_test, verbose=0)
print(f"Test loss: {loss:.4f}, accuracy: {acc:.4f}")

y_pred_probs = best_model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=['NORMAL','PNEUMONIA'],
            yticklabels=['NORMAL','PNEUMONIA'])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

print(classification_report(y_test, y_pred, target_names=['NORMAL', 'PNEUMONIA']))

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.legend()
plt.title('Accuracy')

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend()
plt.title('Loss')
plt.show()
```
