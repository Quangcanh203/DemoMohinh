HUẤN LUYỆN GIAI ĐOẠN 1 HUẤN LUYỆN CHỈ TẦNG HEAD (WARM-UP)


```python
# -*- coding: utf-8 -*-
"""
Mã tinh chỉnh ResNet50 để phân loại X-quang ngực với độ chính xác cải thiện, bao gồm hiển thị ảnh mẫu
"""

from google.colab import drive
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.utils import class_weight
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.optimizers import AdamW
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization, Input
from tensorflow.keras.regularizers import l2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2

# Cấu hình
DEBUG = False
IMG_SIZE = 224
NUM_CLASSES = 2
CLASS_LABELS = ['BÌNH THƯỜNG', 'VIÊM PHỔI']
BATCH_SIZE = 16
EPOCHS_WARMUP = 25
EPOCHS_FINETUNE = 40

# 1. Gắn Google Drive
drive.mount('/content/gdrive', force_remount=True)

# 2. Xác định đường dẫn và tham số
BASE_DRIVE_PATH = '/content/gdrive/My Drive/chest_xray'
NEW_DATASET_ROOT_NAME = 'chest_xray_phan_chia_moi'
NUMPY_DATA_DIR_NAME = 'numpy_data'
MODEL_SAVE_SUBDIR = 'resnet50_trained_models_v5'

numpy_data_dir = os.path.join(BASE_DRIVE_PATH, NEW_DATASET_ROOT_NAME, NUMPY_DATA_DIR_NAME)
model_save_path_root = os.path.join(BASE_DRIVE_PATH, NEW_DATASET_ROOT_NAME, MODEL_SAVE_SUBDIR)

if not os.path.exists(model_save_path_root):
    os.makedirs(model_save_path_root)
    print(f"Đã tạo thư mục lưu mô hình: {model_save_path_root}")
else:
    print(f"Thư mục lưu mô hình đã tồn tại: {model_save_path_root}")

# 3. Hàm tiền xử lý ảnh
def preprocess_images(images, target_size=(IMG_SIZE, IMG_SIZE), apply_blur=True, blur_ksize=(5,5), apply_brightness=True, brightness_factor=1.2, apply_sharpen=True):
    processed_images = []
    for img in images:
        img_resized = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)
        if len(img_resized.shape) == 2:
            img_resized = cv2.cvtColor(img_resized, cv2.COLOR_GRAY2RGB)
        if apply_blur:
            img_blurred = cv2.GaussianBlur(img_resized, blur_ksize, 0)
        else:
            img_blurred = img_resized
        if apply_brightness:
            img_bright = cv2.convertScaleAbs(img_blurred, alpha=brightness_factor, beta=0)
        else:
            img_bright = img_blurred
        if apply_sharpen:
            kernel = np.array([[0, -1, 0],
                               [-1, 5, -1],
                               [0, -1, 0]])
            img_sharp = cv2.filter2D(img_bright, -1, kernel)
        else:
            img_sharp = img_bright
        processed_images.append(img_sharp)
    return np.array(processed_images)

# 4. Hàm hiển thị ảnh mẫu trước và sau chuẩn hóa
def show_sample_images(before_norm, after_norm, num_images=2):
    plt.figure(figsize=(12, 5))
    for i in range(num_images):
        plt.subplot(2, num_images, i + 1)
        plt.imshow(before_norm[i].astype(np.uint8))
        plt.title(f"Before normalization #{i+1}")
        plt.axis('off')
        plt.subplot(2, num_images, num_images + i + 1)
        img_show = after_norm[i]
        plt.imshow(img_show)
        plt.title(f"After normalization #{i+1}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# 5. Tải dữ liệu
print(f"\nĐang tải dữ liệu NumPy từ: {numpy_data_dir}")
try:
    X_train_orig = np.load(os.path.join(numpy_data_dir, "X_train.npy"))
    y_train_orig = np.load(os.path.join(numpy_data_dir, "y_train.npy"))
    X_val_orig = np.load(os.path.join(numpy_data_dir, "X_val.npy"))
    y_val_orig = np.load(os.path.join(numpy_data_dir, "y_val.npy"))
    X_test_final = np.load(os.path.join(numpy_data_dir, "X_test.npy"))
    y_test_final = np.load(os.path.join(numpy_data_dir, "y_test.npy"))
except FileNotFoundError as e:
    print(f"LỖI: Không tìm thấy tệp dữ liệu NumPy.")
    print(e)
    exit()

print(f"X_train_orig: {X_train_orig.shape}, y_train_orig: {y_train_orig.shape}")
print(f"X_val_orig: {X_val_orig.shape}, y_val_orig: {y_val_orig.shape}")
print(f"X_test_final: {X_test_final.shape}, y_test_final: {y_test_final.shape}")

# 6. Tiền xử lý ảnh
print("\n--- TIỀN XỬ LÝ ẢNH: Resize, Blur, Brightness ---")
X_train_preprocessed = preprocess_images(X_train_orig, apply_sharpen=True)
X_val_preprocessed = preprocess_images(X_val_orig, apply_sharpen=True)
X_test_preprocessed = preprocess_images(X_test_final, apply_sharpen=True)
print("Ảnh đã được resize, làm mờ và tăng sáng.")

# 7. Hiển thị ảnh mẫu
print("\n--- HIỂN THỊ ẢNH MẪU TRƯỚC VÀ SAU CHUẨN HÓA ---")
# Chuẩn hóa đơn giản để hiển thị (chia cho 255.0) chỉ cho mục đích trực quan
show_sample_images(X_train_preprocessed, X_train_preprocessed.astype("float32") / 255.0)
print("--- TIẾP TỤC BƯỚC SAU ---")

# 8. Tăng cường dữ liệu và chuẩn hóa ImageNet
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.1,
    brightness_range=[0.8, 1.2],
    preprocessing_function=tf.keras.applications.resnet50.preprocess_input
)

# 9. Chuẩn bị dữ liệu
X_train_orig = X_train_preprocessed
X_val_orig = X_val_preprocessed
X_test_final = X_test_preprocessed

# Gộp dữ liệu huấn luyện và xác thực cho K-Fold
X_for_kfold_and_warmup = np.concatenate((X_train_orig, X_val_orig), axis=0)
y_for_kfold_and_warmup = np.concatenate((y_train_orig, y_val_orig), axis=0)
print(f"\nDữ liệu gộp: X_for_kfold_and_warmup: {X_for_kfold_and_warmup.shape}, y_for_kfold_and_warmup: {y_for_kfold_and_warmup.shape}")

# Chia dữ liệu cho khởi động
X_train_warmup, X_val_warmup, y_train_warmup, y_val_warmup = train_test_split(
    X_for_kfold_and_warmup, y_for_kfold_and_warmup,
    test_size=0.2,
    random_state=42,
    stratify=y_for_kfold_and_warmup
)
print(f"\nDữ liệu khởi động: X_train_warmup: {X_train_warmup.shape}, X_val_warmup: {X_val_warmup.shape}")

# Tính trọng số lớp
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train_warmup),
    y=y_train_warmup
)
class_weight_dict = dict(enumerate(class_weights))

# 10. Tạo mô hình
def create_resnet_model(input_shape=(IMG_SIZE, IMG_SIZE, 3), num_classes=NUM_CLASSES):
    input_tensor = Input(shape=input_shape, name="input_layer_explicit")
    base_model = ResNet50(weights='imagenet', include_top=False, input_tensor=input_tensor, pooling=None)
    base_model.trainable = False
    x = base_model.output
    x = GlobalAveragePooling2D(name="global_avg_pooling")(x)
    x = BatchNormalization(name="bn_after_gap_1")(x)
    x = Dense(1024, activation='relu', kernel_regularizer=l2(0.0005), name="dense_1024")(x)
    x = Dropout(0.4, name="dropout_1")(x)
    x = BatchNormalization(name="bn_after_dense_1")(x)
    x = Dense(512, activation='relu', kernel_regularizer=l2(0.0005), name="dense_512")(x)
    x = Dropout(0.4, name="dropout_2")(x)
    outputs = Dense(num_classes, activation='softmax', name="output_dense")(x)
    return Model(inputs=input_tensor, outputs=outputs)

# 11. Vẽ biểu đồ lịch sử huấn luyện
def plot_training_history(history, title_prefix=""):
    # Kiểm tra xem history là đối tượng History hay từ điển
    if hasattr(history, 'history'):
        history_dict = history.history
    else:
        history_dict = history

    acc = history_dict['accuracy']
    val_acc = history_dict['val_accuracy']
    loss = history_dict['loss']
    val_loss = history_dict['val_loss']
    epochs_range = range(len(acc))
    plt.figure(figsize=(18, 6))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, 'royalblue', label='Độ chính xác huấn luyện')
    plt.plot(epochs_range, val_acc, 'darkorange', linestyle='--', label='Độ chính xác xác thực')
    plt.xlabel('Epoch')
    plt.ylabel('Độ chính xác')
    plt.title(f'{title_prefix} Độ chính xác')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, 'royalblue', label='Hàm mất mát huấn luyện')
    plt.plot(epochs_range, val_loss, 'darkorange', linestyle='--', label='Hàm mất mát xác thực')
    plt.xlabel('Epoch')
    plt.ylabel('Hàm mất mát')
    plt.title(f'{title_prefix} Hàm mất mát')
    plt.legend(loc='upper right')
    plt.grid(True)
    plt.suptitle(f'{title_prefix} Lịch sử huấn luyện')
    plt.tight_layout()
    plt.show()

# 12. Callbacks
lr_scheduler_warmup = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6, verbose=1)
early_stopping_warmup = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
lr_scheduler_finetune = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=6, min_lr=1e-7, verbose=1)
early_stopping_finetune = EarlyStopping(monitor='val_loss', patience=12, restore_best_weights=True, verbose=1)

# 13. Giai đoạn khởi động
print("\n--- BẮT ĐẦU GIAI ĐOẠN KHỞI ĐỘNG ---")
model_for_warmup = create_resnet_model()
model_for_warmup.compile(loss='sparse_categorical_crossentropy', optimizer=AdamW(learning_rate=5e-4, weight_decay=0.01), metrics=['accuracy'])

if DEBUG:
    print("Trạng thái trainable trước KHỞI ĐỘNG:")
    for layer in model_for_warmup.layers:
        print(f"  - {layer.name}: {layer.trainable}")
else:
    print("Tóm tắt trạng thái trainable trước KHỞI ĐỘNG:")
    total_layers = len(model_for_warmup.layers)
    trainable_layers = sum(1 for layer in model_for_warmup.layers if layer.trainable)
    print(f" - Tổng số lớp: {total_layers}, Trainable: {trainable_layers}, Không trainable: {total_layers - trainable_layers}")

# Tạo bộ tạo dữ liệu cho khởi động
train_generator = datagen.flow(X_train_warmup, y_train_warmup, batch_size=BATCH_SIZE)
val_generator = datagen.flow(X_val_warmup, y_val_warmup, batch_size=BATCH_SIZE, shuffle=False)

history_warmup = model_for_warmup.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS_WARMUP,
    callbacks=[lr_scheduler_warmup, early_stopping_warmup],
    class_weight=class_weight_dict,
    verbose=1
)

warmup_model_save_path = os.path.join(model_save_path_root, "ResNet50_WarmUp_ChestXRay_v5.keras")
model_for_warmup.save(warmup_model_save_path)
print(f"Đã lưu mô hình khởi động tại: {warmup_model_save_path}")

print("\nHiển thị lịch sử huấn luyện khởi động:")
plot_training_history(history_warmup, title_prefix="Giai đoạn Khởi động")

# 14. Tinh chỉnh với K-Fold
print("\n--- BẮT ĐẦU GIAI ĐOẠN TINH CHỈNH VỚI K-FOLD ---")
k_folds = 5
skf = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=42)
all_fold_histories = []
fold_no = 1

for train_idx, val_idx in skf.split(X_for_kfold_and_warmup, y_for_kfold_and_warmup):
    print(f'\n========== Fold {fold_no}/{k_folds} ==========')
    X_train_fold, X_val_fold = X_for_kfold_and_warmup[train_idx], X_for_kfold_and_warmup[val_idx]
    y_train_fold, y_val_fold = y_for_kfold_and_warmup[train_idx], y_for_kfold_and_warmup[val_idx]
    print(f"Fold {fold_no}: X_train_fold: {X_train_fold.shape}, X_val_fold: {X_val_fold.shape}")

    # Tải mô hình khởi động
    model_for_finetune = tf.keras.models.load_model(warmup_model_save_path)

    # Giai đoạn 1: Tinh chỉnh conv5_block
    idx_before_gap = -1
    for i, layer in enumerate(model_for_finetune.layers):
        if layer.name == "global_avg_pooling":
            idx_before_gap = i - 1
            break
    if idx_before_gap == -1:
        raise ValueError("Không tìm thấy lớp 'global_avg_pooling'.")

    fine_tune_start_index = -1
    for i, layer in enumerate(model_for_finetune.layers):
        if layer.name.startswith("conv5_block"):
            if fine_tune_start_index == -1:
                fine_tune_start_index = i
                print(f"Giai đoạn 1: Tinh chỉnh từ lớp: {layer.name} (chỉ số {i})")
            if i <= idx_before_gap:
                layer.trainable = not isinstance(layer, tf.keras.layers.BatchNormalization)
        elif i <= idx_before_gap:
            layer.trainable = False
        else:
            layer.trainable = not isinstance(layer, tf.keras.layers.BatchNormalization) or layer.name in ["bn_after_gap_1", "bn_after_dense_1"]

    model_for_finetune.compile(loss='sparse_categorical_crossentropy', optimizer=AdamW(learning_rate=5e-6, weight_decay=0.01), metrics=['accuracy'])

    if DEBUG:
        print(f"Trạng thái trainable Fold {fold_no} (Giai đoạn 1):")
        for i, layer in enumerate(model_for_finetune.layers):
            print(f" - {layer.name}: {layer.trainable} (ResNet: {i <= idx_before_gap})")
    else:
        print(f"Tóm tắt trạng thái trainable Fold {fold_no} (Giai đoạn 1):")
        total_layers = len(model_for_finetune.layers)
        trainable_layers = sum(1 for layer in model_for_finetune.layers if layer.trainable)
        print(f" - Tổng số lớp: {total_layers}, Trainable: {trainable_layers}, Không trainable: {total_layers - trainable_layers}")

    # Bộ tạo dữ liệu cho tinh chỉnh
    train_generator = datagen.flow(X_train_fold, y_train_fold, batch_size=BATCH_SIZE)
    val_generator = datagen.flow(X_val_fold, y_val_fold, batch_size=BATCH_SIZE, shuffle=False)

    history_fold_stage1 = model_for_finetune.fit(
        train_generator,
        validation_data=val_generator,
        epochs=EPOCHS_FINETUNE // 2,
        callbacks=[lr_scheduler_finetune, early_stopping_finetune],
        class_weight=class_weight_dict,
        verbose=1
    )

    # Giai đoạn 2: Tinh chỉnh conv4_block và conv5_block
    fine_tune_start_index = -1
    for i, layer in enumerate(model_for_finetune.layers):
        if layer.name.startswith("conv4_block"):
            if fine_tune_start_index == -1:
                fine_tune_start_index = i
                print(f"Giai đoạn 2: Tinh chỉnh từ lớp: {layer.name} (chỉ số {i})")
            if i <= idx_before_gap:
                layer.trainable = not isinstance(layer, tf.keras.layers.BatchNormalization)
        elif i <= idx_before_gap and not layer.name.startswith("conv5_block"):
            layer.trainable = False
        else:
            layer.trainable = not isinstance(layer, tf.keras.layers.BatchNormalization) or layer.name in ["bn_after_gap_1", "bn_after_dense_1"]

    model_for_finetune.compile(loss='sparse_categorical_crossentropy', optimizer=AdamW(learning_rate=2e-6, weight_decay=0.01), metrics=['accuracy'])

    if DEBUG:
        print(f"Trạng thái trainable Fold {fold_no} (Giai đoạn 2):")
        for i, layer in enumerate(model_for_finetune.layers):
            print(f" - {layer.name}: {layer.trainable} (ResNet: {i <= idx_before_gap})")
    else:
        print(f"Tóm tắt trạng thái trainable Fold {fold_no} (Giai đoạn 2):")
        total_layers = len(model_for_finetune.layers)
        trainable_layers = sum(1 for layer in model_for_finetune.layers if layer.trainable)
        print(f" - Tổng số lớp: {total_layers}, Trainable: {trainable_layers}, Không trainable: {total_layers - trainable_layers}")

    history_fold_stage2 = model_for_finetune.fit(
        train_generator,
        validation_data=val_generator,
        epochs=EPOCHS_FINETUNE // 2,
        callbacks=[lr_scheduler_finetune, early_stopping_finetune],
        class_weight=class_weight_dict,
        verbose=1
    )

    # Kết hợp lịch sử
    history_fold = {}
    for key in history_fold_stage1.history:
        history_fold[key] = history_fold_stage1.history[key] + history_fold_stage2.history[key]
    all_fold_histories.append(history_fold)

    fold_model_save_path = os.path.join(model_save_path_root, f"ResNet50_FineTune_ChestXRay_Fold{fold_no}_v5.keras")
    model_for_finetune.save(fold_model_save_path)
    print(f"Đã lưu mô hình Fold {fold_no} tại: {fold_model_save_path}")

    # Chuẩn hóa dữ liệu trước khi dự đoán
    X_val_fold_preprocessed = tf.keras.applications.resnet50.preprocess_input(X_val_fold.copy())
    y_pred_val_fold_probs = model_for_finetune.predict(X_val_fold_preprocessed)
    y_pred_val_fold = np.argmax(y_pred_val_fold_probs, axis=1)
    val_accuracy_fold = accuracy_score(y_val_fold, y_pred_val_fold)
    print(f'Fold {fold_no} - Độ chính xác xác thực: {val_accuracy_fold:.4f}')

    print(f"\nHiển thị lịch sử huấn luyện cho Fold {fold_no}:")
    plot_training_history(history_fold, title_prefix=f"Tinh chỉnh Fold {fold_no}")

    cm_val_fold = confusion_matrix(y_val_fold, y_pred_val_fold)
    plt.figure(figsize=(7, 5))
    sns.heatmap(cm_val_fold, annot=True, fmt='d', cmap='Blues', xticklabels=CLASS_LABELS, yticklabels=CLASS_LABELS)
    plt.xlabel('Dự đoán')
    plt.ylabel('Thực tế')
    plt.title(f'Ma trận nhầm lẫn - Fold {fold_no}')
    plt.show()
    print(f"Báo cáo phân loại - Fold {fold_no}:")
    print(classification_report(y_val_fold, y_pred_val_fold, target_names=CLASS_LABELS, zero_division=0))

    fold_no += 1

print("\nĐã hoàn thành huấn luyện K-Fold!")

# 15. Chọn và đánh giá mô hình tốt nhất
if all_fold_histories:
    best_fold_index = np.argmax([np.max(history['val_accuracy']) for history in all_fold_histories])
    best_val_accuracy = np.max(all_fold_histories[best_fold_index]['val_accuracy'])
    best_fold_number = best_fold_index + 1
    print(f"\nFold tốt nhất: Fold {best_fold_number} với Độ chính xác xác thực: {best_val_accuracy:.4f}")
    best_model_path = os.path.join(model_save_path_root, f"ResNet50_FineTune_ChestXRay_Fold{best_fold_number}_v5.keras")

    if os.path.exists(best_model_path):
        print(f"\n--- ĐÁNH GIÁ TẬP KIỂM TRA VỚI FOLD {best_fold_number} ---")
        model_best_fold = tf.keras.models.load_model(best_model_path, compile=True)
        X_test_final_preprocessed = tf.keras.applications.resnet50.preprocess_input(X_test_final.copy())
        y_pred_test_probs = model_best_fold.predict(X_test_final_preprocessed)
        y_pred_test = np.argmax(y_pred_test_probs, axis=1)
        test_accuracy = accuracy_score(y_test_final, y_pred_test)
        print(f'Độ chính xác kiểm tra (Fold {best_fold_number}): {test_accuracy:.4f}')

        cm_test = confusion_matrix(y_test_final, y_pred_test)
        plt.figure(figsize=(7, 5))
        sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues', xticklabels=CLASS_LABELS, yticklabels=CLASS_LABELS)
        plt.xlabel('Dự đoán')
        plt.ylabel('Thực tế')
        plt.title(f'Ma trận nhầm lẫn - Tập kiểm tra (Fold {best_fold_number})')
        plt.show()
        print(f"Báo cáo phân loại - Tập kiểm tra (Fold {best_fold_number}):")
        print(classification_report(y_test_final, y_pred_test, target_names=CLASS_LABELS, zero_division=0))
    else:
        print(f"LỖI: Không tìm thấy mô hình tại: {best_model_path}")
else:
    print("Không có lịch sử huấn luyện K-Fold.")

print("\n--- QUÁ TRÌNH HUẤN LUYỆN VÀ ĐÁNH GIÁ HOÀN TẤT ---")
```
