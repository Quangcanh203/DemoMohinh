```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils import class_weight

# 0. Cấu hình và các hằng số
#-------------------------------
print(f"Phiên bản TensorFlow đang sử dụng: {tf.__version__}")

BASE_PATH_PROCESSED_DATA = '/content/gdrive/My Drive/chest_xray/chest_xray_phan_chia_moi'
NUMPY_SAVE_DIR_NAME = 'numpy_data'
NUMPY_DATA_DIR = os.path.join(BASE_PATH_PROCESSED_DATA, NUMPY_SAVE_DIR_NAME)
MODEL_SAVE_PATH = 'best_vit_chest_xray_model_v4_patch16.keras' # Phiên bản mới với patch 16

IMG_SIZE = 224
NUM_CLASSES = 2

# --- Siêu tham số ViT: THỬ NGHIỆM VỚI PATCH_SIZE = 16 ---
PATCH_SIZE = 16 # THAY ĐỔI QUAN TRỌNG
PROJECTION_DIM = 256
TRANSFORMER_LAYERS = 6
NUM_HEADS = 8
MLP_DIM_TRANSFORMER_ENCODER = PROJECTION_DIM * 4 # 1024

NUM_PATCHES = (IMG_SIZE // PATCH_SIZE) ** 2
print(f"Calculated NUM_PATCHES: {NUM_PATCHES}") # Sẽ là 196


# --- Siêu tham số huấn luyện ---
LEARNING_RATE = 5e-5
WEIGHT_DECAY = 5e-5
# **LƯU Ý BATCH_SIZE**: Với PATCH_SIZE=16, số lượng patches tăng lên nhiều.
# Nếu gặp lỗi Out-of-Memory (OOM), HÃY GIẢM BATCH_SIZE xuống 16 hoặc 8.
BATCH_SIZE = 16 # Giảm từ 32 xuống 16 để thử nghiệm trước với patch_size=16
EPOCHS = 100
PATIENCE_EARLY_STOPPING = 20 # Có thể giảm xuống 15 nếu muốn thử nghiệm nhanh hơn

# Tùy chọn Data Augmentation
APPLY_DATA_AUGMENTATION = True

# Tùy chọn Class Weights
APPLY_CLASS_WEIGHTS = True
MANUAL_CLASS_WEIGHTS = False
CUSTOM_WEIGHT_NORMAL = 0.8
CUSTOM_WEIGHT_PNEUMONIA = 1.2

# Tùy chọn Regularization
LABEL_SMOOTHING = 0.0 # Vẫn tắt hẳn do lỗi TypeError persistante

# 1. Mount Google Drive (Giữ nguyên)
#-------------------------------
try:
    from google.colab import drive
    drive.mount('/content/gdrive', force_remount=True)
    print(f"Đã mount Google Drive. Thư mục dữ liệu NumPy dự kiến: {NUMPY_DATA_DIR}")
    if 'google.colab' in str(get_ipython()):
        MODEL_SAVE_PATH = os.path.join('/content', MODEL_SAVE_PATH)
    print(f"Mô hình tốt nhất sẽ được lưu tại: {MODEL_SAVE_PATH}")
except Exception as e:
    print(f"Lỗi khi mount Google Drive hoặc xác định get_ipython: {e}")
    print("Giả sử dữ liệu đã có sẵn tại NUMPY_DATA_DIR nếu không chạy trên Colab.")
    try:
        if 'google.colab' not in str(get_ipython()):
             MODEL_SAVE_PATH = os.path.join(os.getcwd(), MODEL_SAVE_PATH)
    except NameError:
        MODEL_SAVE_PATH = os.path.join(os.getcwd(), MODEL_SAVE_PATH)
    print(f"Mô hình tốt nhất sẽ được lưu tại: {MODEL_SAVE_PATH}")

# 2. Tải dữ liệu NumPy (Giữ nguyên)
#-------------------------------
print("\n--- TẢI DỮ LIỆU NUMPY ---")
if not os.path.exists(NUMPY_DATA_DIR):
    print(f"LỖI: Thư mục {NUMPY_DATA_DIR} không tồn tại. Vui lòng kiểm tra lại đường dẫn.")
    exit()
X_train = np.load(os.path.join(NUMPY_DATA_DIR, "X_train.npy"))
y_train = np.load(os.path.join(NUMPY_DATA_DIR, "y_train.npy"))
X_val = np.load(os.path.join(NUMPY_DATA_DIR, "X_val.npy"))
y_val = np.load(os.path.join(NUMPY_DATA_DIR, "y_val.npy"))
X_test = np.load(os.path.join(NUMPY_DATA_DIR, "X_test.npy"))
y_test = np.load(os.path.join(NUMPY_DATA_DIR, "y_test.npy"))
print(f"Shape của X_train: {X_train.shape}, Shape của y_train: {y_train.shape}")
print(f"Shape của X_val: {X_val.shape}, Shape của y_val: {y_val.shape}")
print(f"Shape của X_test: {X_test.shape}, Shape của y_test: {y_test.shape}")

# 3. Tiền xử lý dữ liệu (Giữ nguyên)
#-------------------------------
print("\n--- TIỀN XỬ LÝ DỮ LIỆU ---")
X_train = X_train.astype("float32") / 255.0
X_val = X_val.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0
print("Đã chuẩn hóa giá trị pixel.")

# 4. Xây dựng các thành phần của Vision Transformer (Giữ nguyên, PatchEncoder sẽ tự điều chỉnh với num_patches mới)
#----------------------------------------------------
print("\n--- XÂY DỰNG CÁC THÀNH PHẦN VIT ---")
class Patches(layers.Layer):
    def __init__(self, patch_size, name=None, **kwargs):
        super().__init__(name=name, **kwargs)
        self.patch_size = patch_size
    def call(self, images):
        batch_size = tf.shape(images)[0]
        patches_data = tf.image.extract_patches(
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1, 1, 1, 1],
            padding="VALID",
        )
        patch_dims = patches_data.shape[-1]
        patches_data = tf.reshape(patches_data, [batch_size, -1, patch_dims])
        return patches_data
    def get_config(self):
        config = super().get_config()
        config.update({"patch_size": self.patch_size})
        return config

class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim, name=None, **kwargs):
        super().__init__(name=name, **kwargs)
        self.num_patches = num_patches
        self.projection = layers.Dense(units=projection_dim)
        self.position_embedding = layers.Embedding(
            input_dim=self.num_patches, output_dim=projection_dim # Sử dụng self.num_patches
        )
    def call(self, patch):
        positions = tf.range(start=0, limit=self.num_patches, delta=1)
        projected_patch = self.projection(patch)
        encoded = projected_patch + self.position_embedding(positions)
        return encoded
    def get_config(self):
        config = super().get_config()
        config.update({
            "num_patches": self.num_patches,
            "projection_dim": self.projection.units
        })
        return config

def mlp_block(x, hidden_units, dropout_rate, transformer_layer_index, block_name_prefix="mlp_block"):
    for units_idx, units in enumerate(hidden_units):
        x = layers.Dense(units, activation=tf.nn.gelu,
                         name=f"{block_name_prefix}_transformer_{transformer_layer_index}_dense_{units_idx}")(x)
        x = layers.Dropout(dropout_rate,
                           name=f"{block_name_prefix}_transformer_{transformer_layer_index}_dropout_{units_idx}")(x)
    return x

# 5. Xây dựng mô hình ViT hoàn chỉnh (Giữ nguyên logic, các tham số sẽ được truyền vào)
#------------------------------------
print("\n--- XÂY DỰNG MÔ HÌNH VIT HOÀN CHỈNH ---")
def create_vit_classifier(input_shape,
                          patch_size,
                          num_patches,
                          projection_dim,
                          transformer_layers,
                          num_heads,
                          mlp_dim_transformer_encoder,
                          num_classes,
                          apply_data_augmentation=False,
                          mlp_head_units_config=None
                         ):
    inputs = layers.Input(shape=input_shape, name="input_image")
    augmented = inputs
    if apply_data_augmentation:
        data_augmentation_layers = keras.Sequential(
            [
                layers.RandomFlip("horizontal", name="aug_flip"),
                layers.RandomRotation(factor=0.05, name="aug_rotate"),
                layers.RandomZoom(height_factor=0.05, width_factor=0.05, name="aug_zoom"),
            ], name="data_augmentation_block",
        )
        augmented = data_augmentation_layers(inputs)

    created_patches = Patches(patch_size, name="patches_extraction_layer")(augmented)
    # num_patches sẽ được tính toán lại ở đầu file và truyền vào đây
    encoded_patches = PatchEncoder(num_patches, projection_dim, name="patch_sequence_encoder_layer")(created_patches)

    for i in range(transformer_layers):
        x1 = layers.LayerNormalization(epsilon=1e-6, name=f"transformer_block_{i}_norm1")(encoded_patches)
        key_dim = projection_dim // num_heads
        if projection_dim % num_heads != 0:
            raise ValueError(f"projection_dim ({projection_dim}) phải chia hết cho num_heads ({num_heads}).")
        attention_output = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=key_dim, dropout=0.1, name=f"transformer_block_{i}_mha"
        )(x1, x1)
        x2 = layers.Add(name=f"transformer_block_{i}_add1")([attention_output, encoded_patches])
        x3 = layers.LayerNormalization(epsilon=1e-6, name=f"transformer_block_{i}_norm2")(x2)
        x3 = mlp_block(x3, hidden_units=[mlp_dim_transformer_encoder, projection_dim],
                       dropout_rate=0.1, transformer_layer_index=i, block_name_prefix="encoder_mlp")
        encoded_patches = layers.Add(name=f"transformer_block_{i}_add2")([x3, x2])

    representation = layers.LayerNormalization(epsilon=1e-6, name="pre_head_norm")(encoded_patches)
    representation = layers.GlobalAveragePooling1D(name="global_avg_pool")(representation)
    features = representation

    if mlp_head_units_config is None:
        mlp_head_units_final = [projection_dim * 2, projection_dim]
    else:
        mlp_head_units_final = mlp_head_units_config

    for idx, units in enumerate(mlp_head_units_final):
        features = layers.Dense(units, activation=tf.nn.gelu, name=f"classification_head_dense_{idx}")(features)
        features = layers.Dropout(0.3, name=f"classification_head_dropout_{idx}")(features)

    logits = layers.Dense(num_classes, activation="softmax", name="output_logits")(features)
    model = keras.Model(inputs=inputs, outputs=logits, name="vit_classifier_patch16") # Đổi tên model
    return model

mlp_head_config = [MLP_DIM_TRANSFORMER_ENCODER, PROJECTION_DIM] # [1024, 256]

vit_classifier = create_vit_classifier(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    patch_size=PATCH_SIZE, # Sẽ là 16
    num_patches=NUM_PATCHES, # Sẽ là 196
    projection_dim=PROJECTION_DIM,
    transformer_layers=TRANSFORMER_LAYERS,
    num_heads=NUM_HEADS,
    mlp_dim_transformer_encoder=MLP_DIM_TRANSFORMER_ENCODER,
    num_classes=NUM_CLASSES,
    apply_data_augmentation=APPLY_DATA_AUGMENTATION,
    mlp_head_units_config=mlp_head_config
)
vit_classifier.summary(line_length=120)

# 6. Biên dịch mô hình (Giữ nguyên)
#-------------------------------
print("\n--- BIÊN DỊCH MÔ HÌNH ---")
num_train_samples = len(X_train)
# steps_per_epoch sẽ thay đổi nếu BATCH_SIZE thay đổi
steps_per_epoch = num_train_samples // BATCH_SIZE
if steps_per_epoch == 0: steps_per_epoch = 1
total_decay_steps = EPOCHS * steps_per_epoch
print(f"Số bước mỗi epoch: {steps_per_epoch} (do BATCH_SIZE={BATCH_SIZE})")
print(f"Tổng số bước suy giảm LR (decay_steps): {total_decay_steps}")

lr_schedule = tf.keras.optimizers.schedules.CosineDecay(
    initial_learning_rate=LEARNING_RATE,
    decay_steps=total_decay_steps,
    alpha=0.0
)
optimizer = tf.keras.optimizers.AdamW(
    learning_rate=lr_schedule, weight_decay=WEIGHT_DECAY
)

if LABEL_SMOOTHING > 0:
    try:
        print(f"ĐANG THỬ sử dụng Label Smoothing với giá trị: {LABEL_SMOOTHING} (TensorFlow {tf.__version__})")
        loss_function = tf.keras.losses.SparseCategoricalCrossentropy(
            from_logits=False,
            label_smoothing=LABEL_SMOOTHING
        )
        _ = tf.keras.losses.SparseCategoricalCrossentropy(label_smoothing=LABEL_SMOOTHING)
        print(f"Label Smoothing ĐƯỢC KÍCH HOẠT với giá trị: {LABEL_SMOOTHING}")
    except Exception as e:
        print(f"LỖI khi cố gắng kích hoạt Label Smoothing với TensorFlow {tf.__version__}: {e}")
        print("Label smoothing sẽ được bỏ qua. Sử dụng loss không có label smoothing.")
        loss_function = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False)
else:
    print(f"Label smoothing KHÔNG được kích hoạt (LABEL_SMOOTHING = {LABEL_SMOOTHING}).")
    loss_function = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False)

vit_classifier.compile(
    optimizer=optimizer,
    loss=loss_function,
    metrics=[keras.metrics.SparseCategoricalAccuracy(name="accuracy")],
)

# Callbacks (Giữ nguyên)
early_stopping = keras.callbacks.EarlyStopping(
    monitor='val_accuracy', patience=PATIENCE_EARLY_STOPPING, restore_best_weights=True, verbose=0
)
model_checkpoint = keras.callbacks.ModelCheckpoint(
    filepath=MODEL_SAVE_PATH, monitor='val_accuracy', save_best_only=True, save_weights_only=False, verbose=0
)
callbacks_list = [early_stopping, model_checkpoint]

# 7. Huấn luyện mô hình (Giữ nguyên)
#-------------------------------
print("\n--- HUẤN LUYỆN MÔ HÌNH ---")
class_weight_dictionary = None
if APPLY_CLASS_WEIGHTS:
    if MANUAL_CLASS_WEIGHTS:
        class_weight_dictionary = {0: CUSTOM_WEIGHT_NORMAL, 1: CUSTOM_WEIGHT_PNEUMONIA}
        print(f"Sử dụng trọng số lớp thủ công: {class_weight_dictionary}")
    else:
        classes_present, counts = np.unique(y_train, return_counts=True)
        print(f"Phân bổ lớp trong tập huấn luyện: {dict(zip(classes_present, counts))}")
        class_weights_calculated = class_weight.compute_class_weight(
            class_weight='balanced', classes=classes_present, y=y_train.flatten()
        )
        class_weight_dictionary = dict(zip(classes_present, class_weights_calculated))
        print(f"Sử dụng trọng số lớp tự động ('balanced'): {class_weight_dictionary}")

history = vit_classifier.fit(
    X_train, y_train,
    batch_size=BATCH_SIZE, # Sẽ là 16
    epochs=EPOCHS,
    validation_data=(X_val, y_val),
    callbacks=callbacks_list,
    class_weight=class_weight_dictionary if APPLY_CLASS_WEIGHTS else None,
    verbose=1
)

# 8. Đánh giá mô hình (Giữ nguyên)
#-------------------------------
print("\n--- ĐÁNH GIÁ MÔ HÌNH TRÊN TẬP TEST ---")
print(f"Tải mô hình tốt nhất từ: {MODEL_SAVE_PATH}")
try:
    custom_objects = {'Patches': Patches, 'PatchEncoder': PatchEncoder}
    vit_classifier_best = keras.models.load_model(MODEL_SAVE_PATH, custom_objects=custom_objects)
    print("Đã tải mô hình tốt nhất thành công.")
except Exception as e:
    print(f"Lỗi khi tải mô hình: {e}. Sử dụng mô hình hiện tại trong bộ nhớ.")
    vit_classifier_best = vit_classifier

loss, accuracy = vit_classifier_best.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")

# 9. Dự đoán và Ma trận nhầm lẫn (Giữ nguyên)
#---------------------------------
print("\n--- DỰ ĐOÁN VÀ MA TRẬN NHẦM LẪN ---")
y_pred_proba = vit_classifier_best.predict(X_test, verbose=0)
y_pred = np.argmax(y_pred_proba, axis=1)
idx_to_class = {0: 'NORMAL', 1: 'PNEUMONIA'}
class_names = [idx_to_class[i] for i in sorted(idx_to_class.keys())]
cm = confusion_matrix(y_test, y_pred, labels=np.arange(NUM_CLASSES))
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names,
            annot_kws={"size": 14})
plt.xlabel('Predicted Label', fontsize=14); plt.ylabel('True Label', fontsize=14)
plt.title('Confusion Matrix - ViT (Patch 16)', fontsize=16); plt.xticks(fontsize=12); plt.yticks(fontsize=12) # Cập nhật title
plt.show()
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=class_names, labels=np.arange(NUM_CLASSES)))

# 10. Vẽ đồ thị accuracy và loss (Giữ nguyên)
#----------------------------------
print("\n--- ĐỒ THỊ ACCURACY VÀ LOSS ---")
if history and history.history:
    if 'accuracy' in history.history and 'val_accuracy' in history.history and \
       'loss' in history.history and 'val_loss' in history.history:

        plt.style.use('seaborn-v0_8-whitegrid')
        plt.figure(figsize=(14, 6))
        epochs_ran = len(history.history['accuracy'])
        epoch_ticks_step = max(1, epochs_ran // 10 if epochs_ran > 0 else 1)

        final_ticks = np.array([])
        if epochs_ran > 0:
            valid_epoch_ticks = np.arange(0, epochs_ran, step=epoch_ticks_step)
            if not valid_epoch_ticks.size or (valid_epoch_ticks.size > 0 and valid_epoch_ticks[-1] < epochs_ran -1) :
                 valid_epoch_ticks = np.append(valid_epoch_ticks, epochs_ran -1)
            valid_epoch_ticks = np.unique(valid_epoch_ticks)
            final_ticks = valid_epoch_ticks[valid_epoch_ticks < epochs_ran]
            if not final_ticks.size and epochs_ran > 0:
                final_ticks = np.array([0])

        plt.subplot(1, 2, 1)
        plt.plot(history.history['accuracy'], label='Train Accuracy', color='royalblue', linewidth=2)
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='darkorange', linewidth=2, linestyle='--')
        plt.title('Model Accuracy (Patch 16)', fontsize=16); plt.xlabel('Epoch', fontsize=14); plt.ylabel('Accuracy', fontsize=14) # Cập nhật title
        if final_ticks.size > 0 : plt.xticks(final_ticks, fontsize=12)
        plt.yticks(fontsize=12); plt.legend(fontsize=12); plt.grid(True)

        plt.subplot(1, 2, 2)
        plt.plot(history.history['loss'], label='Train Loss', color='royalblue', linewidth=2)
        plt.plot(history.history['val_loss'], label='Validation Loss', color='darkorange', linewidth=2, linestyle='--')
        plt.title('Model Loss (Patch 16)', fontsize=16); plt.xlabel('Epoch', fontsize=14); plt.ylabel('Loss', fontsize=14) # Cập nhật title
        if final_ticks.size > 0 : plt.xticks(final_ticks, fontsize=12)
        plt.yticks(fontsize=12); plt.legend(fontsize=12); plt.grid(True)
        plt.tight_layout(); plt.show()
    else:
        print("Dữ liệu accuracy hoặc loss không đầy đủ trong history object để vẽ đồ thị.")
else:
    print("Không có dữ liệu history để vẽ đồ thị (có thể do quá trình huấn luyện bị dừng sớm hoặc lỗi).")
print("\n--- HOÀN TẤT ---")
```
